### Exercise 03.1 <span style="color:red"> Plain vanilla option pricing </span>

Black-Scholes theory assumes that the time evolution of the asset price is stochastic and exhibits geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma$.

Black-Scholes <span style="color:blue"> analytic solution </span> for the price at time t<T:
- European Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- European Put-option price:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

where 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

where $d_2 = d_1 - \sigma\sqrt{T-t}$, and where
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
is the cumulative of Gaussian $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
and where
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

By using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

1. <span style="color:red"> By sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$
2. <span style="color:red"> By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

### For comparison, Black-Scholes analytic solution:

```python
import scipy
from scipy.special import erf
from numpy import exp
from numpy.lib.scimath import log, sqrt
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({
    'font.family':'cmr10',
    'mathtext.fontset': 'cm',
    'axes.formatter.use_mathtext': True,
    'axes.axisbelow': True,
    'figure.figsize': [6.0, 6.0],
    'axes.labelsize': 15,
    'xtick.labelsize': 15,
    'ytick.labelsize': 15,
    'font.size':15,
    'savefig.directory':'./Images'
})

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

### Soluzione
Per valutare il prezzo delle **Opzioni Europee di Put/Call** ho implementato un **Moto Browniano Geometrico** che permette di stimare la variazione del prezzo di un'**Obligazione** nel tempo. Il modello permette di studiare l'evoluzione direttamente da un tempo $t=0$ a un tempo $t=T$:
$$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
Come detto anche sopra, il fatto che $W(t)$ sia distribuito normalmente ci permette di spezzare l'evoluzione di $S(t)$ in $N$ sotto-intervalli.

Alla fine del tempo $T$ di interesse, dato un rate di interessi costante $r$ e un prezzo di esercizio $K$, i costi di **Call** e **Put** per l'obligazione sono dati rispettivamente da:
$$S_{Call}=e^{-rT}\max{(0,S(T)-K)}$$
$$S_{Put}=e^{-rT}\max{(0,K-S(T))}$$

Ho quindi implementato delle simulazioni di moti browniani geometrici con passaggio diretto da $t=0$ a $t=T$ oppure spezzandone l'evoluzione in $100$ sotto-intervalli equivalenti. Studiando l'andamento di $10^5$ repliche indipendenti raggruppate in blocchi da $10^2$ ho fatto una stima dei valori $S_{Call}$ e $S_{Put}$ con le relative incertezze.

```python
data=[]
title=['Diretto','Diretto','Discreto','Discreto']
for i in range(0,4):
    data.append(np.loadtxt("./Data"+str(i)+".txt",delimiter='\t',skiprows=1))
```

```python
plt.figure(figsize=(12,10))
for i in range(0,4):
    x = data[i][:,0]
    y = data[i][:,1]
    err = data[i][:,2]
    plt.subplot(2,2,i+1)
    if i%2==0:
        plt.title('Call-Option '+title[i])
        plt.hlines(call,0,x[-1],linestyles='--',color='r',label='BS = '+'{:.3f}'.format(call))
        c, ec = 'b','lightblue'
    else:
        plt.title('Put-Option '+title[i])
        plt.hlines(put,0,x[-1],linestyles='--',color='r',label='BS = '+'{:.3f}'.format(put))
        c, ec = 'g','lightgreen'
    plt.errorbar(x,y,err,color=c,ecolor=ec,label='Simulation',zorder=1)
    plt.ylabel('Price')
    plt.xlabel('Blocks')
    plt.grid()
    plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
