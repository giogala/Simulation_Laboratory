```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python

df = pd.read_csv('Data0.txt',sep='\t')
x = np.arange(len(df))+1
#x*=1000
plt.errorbar(x,df.iloc[:,0],df.iloc[:,1])
plt.xlabel('blocchi')
plt.ylabel('$C$')
plt.grid(True)
plt.show()
```
