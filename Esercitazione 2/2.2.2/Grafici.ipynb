```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python

df = pd.read_csv('Data.txt',sep='\t')
x = np.arange(len(df))+1
plt.errorbar(x,df.iloc[:,0],df.iloc[:,1])
plt.xlabel('#pass')
plt.ylabel('$\sqrt{<r^2>_{RW}}$')
plt.grid(True)
plt.show()
```

```python
dfs = pd.read_csv('Data2.txt',sep='\t')
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.plot3D(dfs.iloc[:,0],dfs.iloc[:,1],dfs.iloc[:,2],"-")
plt.show()
```
