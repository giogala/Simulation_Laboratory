### Exercise 2.1
----

- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

Show a picture of both your estimations of $I$ and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$.

```python
i=np.loadtxt("./2.1/Data.txt",delimiter='\t',skiprows=0)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.errorbar(i[:,0],i[:,1]-1,i[:,2],color='b',ecolor='lightblue')
plt.xlabel('Blocks')
plt.ylabel('$I-1$')
plt.title('Integral')
plt.grid()
plt.subplot(1,2,2)
plt.plot(i[:,0],i[:,2],color='b')
plt.xlabel('Blocks')
plt.ylabel('$\sigma$')
plt.title('Error')
plt.grid()
plt.suptitle('Uniform Sampling')
plt.tight_layout()
plt.show()
```

```python
i2=np.loadtxt("./2.1/Data2.txt",delimiter='\t',skiprows=0)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.errorbar(i2[:,0],i2[:,1]-1,i2[:,2],color='b',ecolor='lightblue')
plt.xlabel('Blocks')
plt.ylabel('$I-1$')
plt.title('Integral')
plt.grid()
plt.subplot(1,2,2)
plt.plot(i2[:,0],i2[:,2],color='b')
plt.xlabel('Blocks')
plt.ylabel('$\sigma$')
plt.title('Error')
plt.grid()
plt.suptitle('Importance Sampling')
plt.tight_layout()
plt.show()
```

### Exercise 2.2
----

- 3D Random Walks (RW) on a cubic lattice and in the continuum: Repeat many times (e.g. say $10^4$) the simulation of a random walk in 3D always starting at the origin

1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$
2. in the continuum; at each discrete time the walker makes a step of length equal to $a(=1)$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

Show a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ for both RWs, with their statistical uncertainties, as function of the step $i\in [0,10^2]$.

Suggestion: divide your $M$ throws into a reasonable number of blocks, compute $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ as function of the step $i\in [0,10^2]$ in each block and use these values (for fixed $i$) to obtain the average value and its statistical uncertainty.

Note that you could try to fit both results with a function like $f(N)=k\sqrt{N}$. Do your results indicate a diffusive behavior?

```python
rms = np.loadtxt("./2.2.1/Data.txt",delimiter='\t',skiprows=1)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.errorbar(rms[:,0],rms[:,1],rms[:,2],color='b',ecolor='lightblue')
plt.ylabel('$\sqrt{\langle r^2\\rangle_{RW}}$')
plt.xlabel('Steps')
plt.title('Displacement')
plt.grid()
plt.subplot(1,2,2)
plt.errorbar(rms[:,0],rms[:,1]-np.sqrt(rms[:,0]),rms[:,2],color='b',ecolor='lightblue')
plt.ylabel('$\sqrt{\langle r^2\\rangle_{RW}}-\sqrt{N_{Steps}}$')
plt.xlabel('Steps')
plt.title('Comparison')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
t = np.loadtxt("./2.2.1/Data2.txt",delimiter='\t',skiprows=1)
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.plot3D(t[:,0],t[:,1],t[:,2],"-o",color='b',linewidth=1,markersize=3)
plt.show()
```
