```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

df = pd.read_csv('Data.txt',sep='\t')
x = np.arange(len(df))+1
x*=1000
plt.errorbar(x,df.iloc[:,0]-0.5,yerr=df.iloc[:,1])
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
dfs = pd.read_csv('Data2.txt',sep='\t')
x = np.arange(len(dfs))+1
x*=1000
plt.errorbar(x,dfs.iloc[:,0]-1/12,yerr=dfs.iloc[:,1])
plt.xlabel('#throws')
plt.ylabel('<(r-1/2)^2>-1/12')
plt.grid(True)
plt.show()
```

```python
dft = pd.read_csv('Data3.txt',sep='\t')
plt.hist(dft)
plt.xlabel('$\chi^2$')
plt.show()
```
